# egmenting and Clustering Neighborhoods in Toronto

# Part I

### import canada postal codes from wikipedia and format it as csv file
I have collected the table from the wiki page by copy and pasting and put the raw data into a csv file

```python
import pandas as pd
```

```python
# read the file
df = pd.read_csv("CanadaPostalCodes.csv")

# set the columns name
df.columns = ['Postcode','Borough','Neighbourhood']

df.head()
```

## Pre processing with the data

```python
# ignore the rows with Not assigned value in Borough colomun
df = df.loc[df['Borough'] != 'Not assigned']

# merge the duplicate postal codes
df = df.groupby('Postcode').agg({'Borough':'first','Neighbourhood': ', '.join}).reset_index()

# set the Neighbourhood value as same as the the Borough if its value is Not assigned
df.loc[df['Neighbourhood'] == 'Not assigned', 'Neighbourhood'] = df['Borough']
```

## View the shape of the dataframe

```python
# view the the dataframe shape
df.shape
```

```python
df
```

# Part II : merging the postal codes with the Longitude and latitude

```python
df_geo = pd.read_csv('Geospatial_Coordinates.csv')

# merge the two dataframe based on the postal code column
df = df.join(df_geo)

df
```

# Part III Visualizing the datafram in a map
import the necesseary libraries

```python
import folium
```

## plot each nieghberhood in the map by looping on each row in the dataframe

```python
# create map of New York using latitude and longitude values
map_newyork = folium.Map(location=[43.70011, -79.4163], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df['Latitude'], df['Longitude'], df['Borough'], df['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_newyork)  
    

map_newyork
```
